# 3 DESTRIPANDO UNA CLASE FRACCION
Durante el desarrollo de esta conferencia vamos a ir evolucionando la definición de una clase `Fraccion` que nos servirá para representar y trabajar con números fraccionarios, es decir aquellos conformados por un numerador y un denominador que serán números enteros.

La definición de la función `__init__` a continuación nos permite crear objetos de tipo `Fraccion` en la forma `Fraccion(3,4)` donde `3` sería el _numerador_ y `4` el _denominador_ o en la forma `Fraccion(5)` donde el numerador es `5` y el denominador (al omitir el segundo parámetro) se asume por defecto que es `1` según la definición def `__init__(self, n: int, d: int=1)`.

Como ya hemos dicho anteriormente la especificación `n: int, d: int` indica que los tipos de los parámetros` n` y `d `es `int`. Esto solo persigue mayor legibilidad en el código. Como puede verse en el código más abajo si omitimos esta especificación el código ejecutará de igual manera.

La inicialización también verifica que el denominador no sea `0` (dando excepción en tal caso) y normaliza el signo para que en caso de ser una fración negativa sea el numerador el que sea negativo.

````python
from lmdb import InvalidParameterError
class Fraccion:
    def __init__(self, n: int, d: int=1):
    # Comente la definición de arriba y descomente esta para que vea que el código ejecuta pro igual
    # def __init__(self, n, d=1):
      # Parámetro d por defecto es 1, de modo que se puede escribir Fraccion(5). Se asume que denomnador es 1
      if (d==0):
        raise InvalidParameterError("denominador no puede ser 0")
      # Normalizar el signo: el denominador siempre positivo
      # Es decir si se escribe Fraccion(1/-2) se asume que es lo mismo que Fraccion(-1,2) y Fraccion(-2,-3) es lo mismo que Fraccion(2,3).
      #
      if d < 0:
        n = -n
        d = -d
      self.num = n
      self.den = d

print(Fraccion(3, 4))
print(Fraccion(12, -4))
print(Fraccion(-21, -35))
print(Fraccion(5))
#No es la forma en que nos gustaría visualizar una Fraccion
````

## 3.1 Mejorando la visualización de una fracción con la función __str__
La ejecución del código anterior nos da una salida. Al hacer print(Fracción(3,4)) Python nos escribe la cadena

`<__main__.Fraccion object at 0x000001ED272F34D0>`

`__main__.Fraccion object `nos dice que es una objeto de tipo `Fraccion` y el valor `0x000001ED272F34D0` es una suerte de posición en memoria de dónde está ubicado el objeto, lo cual realmente no nos aporta mucho. El problema es que Python no puede adivinar qué es lo que realmente se está intentando definir con la clase `Fraccion`.

En el código a continuación se redefine el método especial `__str__` que devuelve una cadena con la forma en que queremos visualizar una fracción. Una fracción como `Fraccion(3,4)` la queremos visualizar como `3/4`, pero cuando tiene denominador con valor` 1` solo queremos visualizar el numerador como entero. Es decir `Fraccion(5)` se escribirá `5`.

_Como se verá en próximas clases cuando se estudie la_ **herencia**, _todas las clases heredan de una clase base_ **object** _en la cual entre otras está la_ _definicón primigenia de algunos métodos como_ ```__str__``` _y que luego cada clase particular podrá redefinir a su convenencia_.

````python
#Mejorando visualización de una Fraccion con __str__
from lmdb import InvalidParameterError
class Fraccion:
    def __init__(self, n: int, d: int=1):
      #Parámetro d por defecto es 1, de modo que se puede escribir Racional(5). Se asume que denomnador es 1
      if (d==0):
        raise InvalidParameterError("denominador no puede ser 0")
      # Normalizar el signo: el denominador siempre positivo
      # Es decir si se escribe Racional(1/-2) se asume que es lo mismo que Racional(-1,2) y Racional(-2,-3) es lo mismo que Racional(2,3).
      #
      if d < 0:
        n = -n
        d = -d
      self.num = n
      self.den = d

    def __str__(self) -> str:
        if self.den == 1: #si el denominador es 1 escribe solo el numerador
            return str(self.num)
        return f'{self.num}/{self.den}'

print(Fraccion(3, 4))
print(Fraccion(12, -4))
print(Fraccion(-21, -35))
print(Fraccion(5))
````

### 3.2.1 La funcion __repr__
Si se redefine la función __repr__ ésta será la que Python aplica cuando quiera ver como se quiere usar la representación del objeto cuando se use dentro de otro. ## 3.3 Simplificando las fracciones
Para simplificar las operaciones aritméticas con fracciones (que veremos en las secciones posteriores), y para evitar numeradores y denominadores muy grandes, cada vez que se intente crear un objeto `Fraccion` vamos a simplificar éste (como mismo solemos hacer en el cálculo manual). Vamos a dividir numerador y denominador entre un entero que los divida a ambos, en este caso entre el _máximo común divisor_ de ambos.

````python
#Simplificando representación. Por ejemplo una fracción como Fraccion(2,4) se representará como Fraccion(1,2)
from lmdb import InvalidParameterError
class Fraccion:
    def __init__(self, n: int, d: int=1):
        def mcd(n: int, m: int) -> int:
            while m != 0:
                (m, n) = (n % m, m)
            return n
        if (d==0):
            raise InvalidParameterError("denominador no puede ser 0")
                # Normalizar el signo: el denominador siempre positivo
        if d < 0:
            n = -n
            d = -d
        #Simplificando la fracción
        max_comun_div = mcd(n, d)
        self.num = int(n/max_comun_div)
        self.den = int(d/max_comun_div)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f'{self.num}/{self.den}'

    def __repr__(self) -> str:
        return self.__str__()

lista_fracciones: list[Fraccion] = [Fraccion(3,4), Fraccion(12, -4), Fraccion(-21, -35), Fraccion(5, 1)]
print(lista_fracciones)
tuplo_fracciones: tuple[Fraccion, Fraccion] = (Fraccion(2, 4), Fraccion(3, 4))
print(tuplo_fracciones)
diccionario_fracciones:dict[str, Fraccion] = {"Racional(1,5)":Fraccion(1, 5), "Racional(4,6)":Fraccion(4, 6)}
print(diccionario_fracciones)
````

Para refrescarle la programación note que se ha incluido la definición una función `mcd` dentro de la función `__init__ `para consumo de ésta. Realmente Python tiene dentro el módulo una función `gcd` (_great common divisor_) que hace esto

````python
#Usando gcd de Python en lugar de mcd
from math import gcd
from lmdb import InvalidParameterError
class Fraccion:
    def __init__(self, n: int, d: int=1):
        # def mcd(n: int, m: int) -> int:
        #     while m != 0:
        #       (m, n) = (n % m, m)
        #     return n
        if (d==0):
            raise InvalidParameterError("denominador no puede ser 0")
                # Normalizar el signo: el denominador siempre positivo
        if d < 0:
            n = -n
            d = -d
        #Simplificando la fracción
        max_comun_div = gcd(n, d)
        self.num = int(n/max_comun_div)
        self.den = int(d/max_comun_div)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f'{self.num}/{self.den}'

    def __repr__(self) -> str:
        return self.__str__()

lista_fracciones: list[Fraccion] = [Fraccion(3,4), Fraccion(12, -4), Fraccion(-21, -35), Fraccion(5, 1)]
print(lista_fracciones)
tuplo_fracciones: tuple[Fraccion, Fraccion] = (Fraccion(2, 4), Fraccion(3, 4))
print(tuplo_fracciones)
diccionario_fracciones:dict[str, Fraccion] = {"Racional(1,5)":Fraccion(1, 5), "Racional(4,6)":Fraccion(4, 6)}
print(diccionario_fracciones)
````

## 3.4 Aritmética de fracciones
Si ejecutamos el código a continuación Python lanzará una excepción porque no sabe cómo sumar dos fracciones. El operador `+` no sabe cómo sumar dos objetos tipo` Fraccion`

`TypeError: unsupported operand type(s) for +: 'Fraccion' and 'Fraccion' `

````python
#Tratando de hacer aritmética de fracciones
from math import gcd
from lmdb import InvalidParameterError
class Fraccion:
    def __init__(self, n: int, d: int=1):
        if (d==0):
            raise InvalidParameterError("denominador no puede ser 0")
                # Normalizar el signo: el denominador siempre positivo
        if d < 0:
            n = -n
            d = -d
        #Simplificando la fracción
        max_comun_div = gcd(n, d)
        self.num = int(n/max_comun_div)
        self.den = int(d/max_comun_div)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f'{self.num}/{self.den}'

    def __repr__(self) -> str:
        return self.__str__()

#Da excepción la suma no está definida para Fraccion
suma:Fraccion = Fraccion(1, 4) + Fraccion(3, 8)
print(suma)
````

Sin embargo, Python tiene también funciones mágicas del estilo `__X__` para cada uno de los operadores aritméticos de suma `+`, resta `-`, multiplicación `*` y división `/` a través de las funciones `__add__`, `__sub__`, `__mult__`, `__truediv__`

Suponemos que el lector repase sus conocimientos de las operaciones aritméticas con números fraccionarios que están implementadas en el código a continuación. Cada operador da como resultado un objeto de tipo `Fraccion`, por lo que al aplicar la simplificación implementada en `__init__` de `Fraccion` nos garantiza que el resultado de una expresión aritmética con fracciones siempre quede simplificado. Note que al ejecutar el código a continuación el resultado de hacer

`print(f"{un_cuarto} + {tres_octavos} * {Fraccion(8, 3)} == {un_cuarto + tres_octavos * Fraccion(8, 3)}")` nos queda como

`1/4 + 3/8 * 8/3 == 5/4 `y no como `30/24`

Esta implementación también nos da un mensaje más elocuente si se comete el error de querer evaluar una expresión como `Fraccion(1,4) + 1`. Al detectar que el operando (el entero `1 `en este caso) que se le pasa a al parámetro `r` en `__add__(self, r: Fraccion)` no es un objeto `Fraccion`. Ver en el código siguiente cómo nos lanzaría la excepción

`TypeError: No se puede sumar una Fraccion con un int`

**NOTA**:

El lector podrá argumentar que el valor entero `n` puede considerarse como la `Fraccion(n,1)` y que entonces Python haga la conversión. Sin embargo, esto no es una práctica recomendada. Si se permite escribir `Fraccion(1,4) + 1 `entonces por qué no permitir también `1 + Fraccion(1,4)` ya que a fin de cuentas la suma `+ `debe ser una operación conmutativa. Pero si se escribe `1+Fraccion(1,4)` Python estaría invocando en todo caso a la funcion `__add__ `de `int `con parámetro de tipo `Fraccion`. PERO NO ESTAMOS DEFINIENDO LA FUNCION `add `de `Fraccion`. Descomente la última línea del códio a continuación y observe lo que ocurre




````python
from math import gcd
from lmdb import InvalidParameterError
#Operadores aritméticos con Fraccion
class Fraccion:
    def __init__(self, n: int, d: int=1):
        if (d==0):
            raise InvalidParameterError("denominador no puede ser 0")
        # Normalizar el signo: el denominador siempre positivo
        if d < 0:
            n = -n
            d = -d
        #Simplificando la fracción usando gcd de Python
        max_comun_div = gcd(n, d)
        self.num = int(n/max_comun_div)
        self.den = int(d/max_comun_div)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f'{self.num}/{self.den}'

    def __repr__(self) -> str:
        return self.__str__()

    def __add__(self, r: Fraccion) -> Fraccion:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede sumar a una Fraccion un {type(r).__name__}")
        num = (self.num * r.den +
               r.num * self.den)
        den = self.den * r.den
        return Fraccion(num, den)

    def __sub__(self, r: Fraccion) -> Fraccion:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede restar a una Fraccion un valor {type(r).__name__}")
        num = (self.num * r.den -
               r.num * self.den)
        den = self.den * r.den
        return Fraccion(num, den)

    def __mul__(self, r: Fraccion) -> Fraccion:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede multiplicar a una Fraccion un valor {type(r).__name__}")
        num = self.num * r.num
        den = self.den * r.den
        return Fraccion(num, den)

    def __truediv__(self, r: Fraccion) -> Fraccion:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede dividir una Fraccion con un {type(r).__name__}")
        if r.num == 0:
            raise InvalidParameterError("No se puede dividir por cero")
        num = self.num * r.den
        den = self.den * r.num
        return Fraccion(num, den)

un_cuarto = Fraccion(1, 4)
tres_octavos = Fraccion(3, 8)
print(f"{un_cuarto} + {tres_octavos} + {Fraccion(4, 5)} == {un_cuarto + tres_octavos + Fraccion(4, 5)}")
print(f"{un_cuarto} + {tres_octavos} * {Fraccion(8, 3)} == {un_cuarto + tres_octavos * Fraccion(8, 3)}")
print(f"{Fraccion(1, 5)} * {Fraccion(4, 5)} == {Fraccion(1, 5) * Fraccion(4, 5)}")
print(f"{Fraccion(2, 7)} / {Fraccion(2, 7)} == {Fraccion(2, 7) / Fraccion(2, 7)}")

#Descomentar para probar el tipo de error que nos dan las operaciones siguientes
# r = Fraccion(1, 5) + 1

# r:Fraccion = 1 + Fraccion(1, 5)
````

## 3.4 Comparacion de Fracciones
### 3.4.1 Comparacion por igualdad ==. La función __eq__

Si comparamos por igualdad `==` dos variables que tienen el mismo objeto de tipo `Fraccion`, como las variables `un_quinto` y `el_mismo_quinto` del código a continuación, la comparación da `True`.

Pero si hacemos la asignación `un_quinto = Fraccion(1,5)` y luego comparamos `un_quinto == Fraccion(1,5)` la comparación da `False` porque se han creado dos objetos `Fraccion` y aunque ambos fueron inicializados con los parámetros `1` y `5` tienen sus propios atributos `num` y `den`. Por defecto la comparación `==` de Python no compara variable de instancia contra variable de instancia de cada objeto.

Sin embargo, si en la clase `Fraccion` implementamos una funcion `__eq__(self, r)` entonces el operador `==` aplicará dicha función cuando se aplique sobre dos operandos `Fraccion` como es el caso en `un_quinto == Fraccion(1,5)`. Descomente en el código a continuación la definición de `__eq__`. Cómo el constructor `__init__ `de `Fraccion` garantiza que la fracción siempre quede _simplificada_ basta entonces con comparar que `num` y `den `de `self `y del parámetro` r` sean respectivamente iguales

````python
from lmdb import InvalidParameterError
from math import gcd
#Comparación con == y con eq
class Fraccion:
    def __init__(self, n: int, d: int=1):
        if (d==0):
            raise InvalidParameterError("denominador no puede ser 0")
        # Normalizar el signo: el denominador siempre positivo
        if d < 0:
            n = -n
            d = -d
        #Simplificando la fracción usando gcd de Python
        max_comun_div = gcd(n, d)
        self.num = int(n/max_comun_div)
        self.den = int(d/max_comun_div)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f'{self.num}/{self.den}'

    def __repr__(self) -> str:
        return self.__str__()

    def __eq__(self, r: Fraccion) -> bool:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede comparar con == una Fraccion con un {type(r).__name__}")
        return self.num == r.num and self.den == r.den

un_quinto = Fraccion(1, 5)
el_mismo_quinto = un_quinto
print(f"un_quinto == el_mismo_quinto es { un_quinto == el_mismo_quinto }")
otro_quinto = Fraccion(1, 5)
dos_decimos = Fraccion(2, 10)
#print(f"un_quinto == Fraccion(1,5) es {un_quinto == 1}")
print(f"un_quinto == Fraccion(2,10) es {un_quinto == Fraccion(2,10)}")

#Pruebe otra vez descomentando el segmento de codigo con la definción de __eq__
````

### 3.4.2 Ampliando la definición de Fraccion con todos los operadores de comparación

Ejecute el código a continuación. Se crea la lista de fracciones

`[Fraccion(1, 5), Fraccion(5, 6), Fraccion(-3, 8), Fraccion(4), Fraccion(2, 21), Fraccion(5, 3)]` cuya visualización sería

`[1/5, 5/6, -3/8, 4, 2/21, 5/3]`

pero luego a esta lista se le quiere aplicar la función` sort(`) con la intención de obtener la lista ordenada. La función `sort` usa los operadores de comparación de orden `<` y otros para valores numéricos, string pero no tiene por qué saber cuál es el orden < para `Fraccion`. Por lo tanto la ejecución del siguiente código dará error cuando `sort` intente comparar si una fracción es menor que otra usando `<`. Se lanza la excepción

`TypeError: '<' not supported between instances of 'Fraccion' and 'Fraccion'`


````python
from lmdb import InvalidParameterError
from math import gcd
from typing import List

#Queriendo ordenar una lista de fracciones
class Fraccion:
    def __init__(self, n: int, d: int=1):
        if (d==0):
            raise InvalidParameterError("denominador no puede ser 0")
        # Normalizar el signo: el denominador siempre positivo
        if d < 0:
          n = -n
          d = -d
        #Simplificando la fracción usando gcd de Python
        max_comun_div = gcd(n, d)
        self.num = int(n/max_comun_div)
        self.den = int(d/max_comun_div)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f'{self.num}/{self.den}'

    def __repr__(self) -> str:
        return self.__str__()

lista_fracciones:List[Fraccion] = [Fraccion(1, 5), Fraccion(5, 6), Fraccion(-3, 8), Fraccion(4), Fraccion(2, 21), Fraccion(5, 3)]
print(lista_fracciones)
print([Fraccion(1, 5), Fraccion(5, 6), Fraccion(-3, 8), Fraccion(4), Fraccion(2, 21), Fraccion(5, 3)].sort())
#Da error porque sorted se basa en el operador < que no está definido para Fraccion
````

El código a continuación implementa la función` __lt__(self, r: Fraccion) -> bool` para comparar si un objeto` Fraccion` es menor que otro (**lt** por las iniciales en inglés de **less than**). Python aplica esta función cuando se usa el operador `<`. En este caso el código

`lista_fracciones.sort()
print(f"lista_racionales luego de sort {lista_fracciones}")` ordena la lista_fracciones y deja el resultado sobre la misma variable que el escribirla con

`print(f"lista_racionales luego de sort {lista_fracciones}")` nos da el resultado

`[-3/8, 2/21, 1/5, 5/6, 5/3, 4]`

En el código a continuación se han definido las funciones `__lt__, __gt__, __le__, __ge__ `correspondientes a los operadores `<, <=, >, >=` aunque la implementación que hace Python de `sort` y `sorted` solo se basan en la implementación de `__lt__`

````python
from lmdb import InvalidParameterError
from math import gcd
from typing import List

#Definiendo todos los operadores de comparación
class Fraccion:
    def __init__(self, n: int, d: int=1):
        if (d==0):
            raise InvalidParameterError("denominador no puede ser 0")
        # Normalizar el signo: el denominador siempre positivo
        if d < 0:
            n = -n
            d = -d
        #Simplificando la fracción usando gcd de Python
        max_comun_div = gcd(n, d)
        self.num = int(n/max_comun_div)
        self.den = int(d/max_comun_div)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f'{self.num}/{self.den}'

    def __repr__(self) -> str:
        return self.__str__()

    def __eq__(self, r: Fraccion) -> bool:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede comparar con == una Fraccion con un {type(r).__name__}")
        return self.num == r.num and self.den == r.den

    def __ne__(self, r: Fraccion) -> bool:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede comparar con != una Fraccion con un {type(r).__name__}")
        return self.num != r.num or self.den != r.den

    def __lt__(self, r: Fraccion) -> bool:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede comparar con < una Fraccion con un {type(r).__name__}")
        return self.num*r.den < r.num*self.den

    def __gt__(self, r: Fraccion) -> bool:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede comparar con > una Fraccion con un {type(r).__name__}")
        return self.num*r.den > r.num*self.den

    def __le__(self, r: Fraccion) -> bool:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede comparar con <= una Fraccion con un {type(r).__name__}")
        return self.num*r.den <= r.num*self.den

    def __ge__(self, r: Fraccion) -> bool:
        if not isinstance(r, Fraccion):
            raise TypeError(f"No se puede comparar con >= una Fraccion con un {type(r).__name__}")
        return self.num*r.den >= r.num*self.den

print(f"{Fraccion(1, 4)} > {Fraccion(3, 5)} es {Fraccion(1, 4) > Fraccion(3, 5)}")
print(f"{Fraccion(1, 4)} <= {Fraccion(1, 4)} es {Fraccion(1, 4) <= Fraccion(1, 4)}")
print(f"{Fraccion(1, 4)} < {Fraccion(3, 5)} es {Fraccion(1, 4) <= Fraccion(3, 5)}")
````
