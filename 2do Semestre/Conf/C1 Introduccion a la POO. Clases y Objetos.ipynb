## 1.1 A la antigua

En el código a continuación tenemos dos variables` p1` y `p2` a las que le asignamos respectivamente dos tuplos formados cada uno por dos enteros que nos representan las coordenadas` x` y `y` de dos puntos. Luego se define una función distancia que recibe dos parámetros que se han llamado punto1 y punto2. La función presupone que a cada parámetro se le pasa un tuplo de dos enteros que son las coordenadas de cada punto y devuelve un float que es la distancia entre los puntos calculada por el teorema de Pitágoras.

Este código de manera directa, salvo lo que se pueda documentar con comentarios, que se está trabajando con un concepto **Punto** el cual se está representando por un tuplo de dos enteros, tampoco nos dice que la función **distancia** está codificada suponiendo que se le invoca con dos parámetros cuyos valores tengan que ser tuplos de enteros que se correspondan con dicha representación de **Punto**  . De modo que estas nociones de punto y distancia quedan tan dispersas en el código como sea la disciplina, claridad y orden que le ponga el programador a la hora de escribir el código.

```python
from math import sqrt

p1 = (0, 0)   # Origen
p2 = (3, 4)   # Punto en el plano

# Función que calcula la distancia euclidiana entre dos puntos
def distancia(punto1, punto2):
    x1, y1 = punto1
    x2, y2 = punto2
    dx = x2 - x1
    dy = y2 - y1
    return sqrt(dx**2 + dy**2)

# Invocar la función
dist = distancia(p1, p2)
print(f"Distancia entre {p1} y {p2}: {dist}")
```

## 1.2 Encapsulamiento con las clases

Sería conveniente poder al menos agrupar visualmente este concepto de **Punto** y las funciones para trabajar con puntos de manera que queden sintácticamente más legibles en el código.

Para hacer esta "agrupación" Python tiene el recurso **class** (clase). La clase permite crear **objetos** (que se denominan **instancias** de la clase. A cada objeto se le pueden asociar variables y asignarles valores. Dentro de la clase se pueden definir funciones que solo se pueden invocar a través de la clase o de instancias de la clase.

En el código a continuación se define una clase `Punto` y dentro de la clase se ha definido una función `distancia`. Esto hace más legible y claro que esta función forma parte de una misma definición y que para invocarla hay que hacerlo a través de la clase, es decir `Punto.distancia(p1,p2)`.


```python
from math import sqrt
class Punto:
    def distancia(punto1, punto2):
        return sqrt((punto1.x-punto2.x)**2 + (punto1.y-punto2.y)**2)

#se crean dos instancias de Punto en las variables p1 y p2 respectivamente
p1 = Punto()
p2 = Punto()
#se definen dos variables x y y tanto en p1 como en p2 y se le asignan valores
p1.x = 0
p1.y = 0
p2.x = 3
p2.y = 4
print(f"Distancia entre {p1} y {p2}: {Punto.distancia(p1,p2)}")
```

Se ha ganado tal vez un poco más de claridad pero no mucho más. La definición de la función distancia presupone que cada parámetro (`punto1` y `punto2`) tiene dos atributos` x` y `y` que se considerarán las coordenadas de un punto para con ello calcular la distancia entre ambos aplicando la fórmula pitagórica. Y esto aunque se puede deducir si se analiza el código, no está especificado de forma clara.

Tampoco se garantiza que una vez creadas dos variables` p1` y `p2` con valores instancias de `Punto` y antes de pretender calcular la distancia se le añadan a ambas dos variables: una variable `x` y una variable `y` que valgan para representar las coordenadas de cada punto.

El siguiente código daría error porque si bien a la variable `p1` se le añadieron dos variables `x` y `y`, a la variable `p2` lo que se añadieron fueron dos variables` a `y `b`. Entonces al ejecutar la llamada a la funcion distancia Python encontrará que el parámetro `punto2` no tiene ninguna variable` x` ni ninguna variable `y` y por tanto da la excepción que se uede apreciar al ejecutar el código siguiente.

Lo que sería deseable es que en la propia definición de la clase `Punto` se pudiera especificar que todas las instancias creadas de una clase `Punto` ya tengan sus correspondientes variables `x` y `y`, en este caso para tomar valores enteros. La solución a esto se verá en la sección siguiente

```python
from math import sqrt
class Punto:
    def distancia(punto1, punto2):
        return sqrt((punto1.x-punto2.x)**2 + (punto1.y-punto2.y)**2)

#se crean dos instancias de Punto en las variables p1 y p2 respectivamente
p1 = Punto()
p2 = Punto()
#Python es indiferente a los atributos que se le asignan a cada una de estas variables
p1.x = 0
p1.y = 0
p2.a = 3 #la ejecución de la función distancia dará error porque el parámetro punto2 no tiene un atributo de nombre x sino de nombre a
p2.b = 4
print(f"Distancia entre {p1} y {p2}: {Punto.distancia(p1,p2)}")
```

### 1.2.1 Constructor. El método especial __init__ y las variables de instancia
Python tiene un método especial `__init__` que se puede definir dentro de cada clase. Debe interpretarse (como su propio nombre lo indica) como una suerte de inicialización de la instancia que se ejecutará al crear una instancia de la clase. Este método puede (de hecho ese es su mayor utilidad) tener parámetros que sirven de información necesaria para inicializar las variables que una instancia debe tener y/o para hacer las verificaciones necesarias para crear correctamente la instancia.

Considere para el ejemplo de código abajo que se ejecuta la acción

`p1 = Punto(0,0)`

Python crea una intancia de `Punto` como valor que asigna a `p1` y se ejecuta una llamada de la forma `Punto.__init__(p1,0,0)`

`__init__` recibe la variable `p1` a través del parámetro `self` de modo que cuando en `__init__` se ejecuta el código `self.x = X` y `self.y = Y` lo que está haciendo es añadiendo las variables` x` y `y` a la variable `p1`. Lo mismo ocurrirá cuando se haga `p2 = Punto(3,4)`.

Con esto se da uniformidad, garantía y legibilidad en el código de que todas los objetos de tipo `Punto` (esto es creados con la clase `Punto`) tengan sus correspondientes componentes `x` y `y`, lo que en la jerga de la POO (Programación Orientada a Objetos) se llaman **variables de instancia** de la clase.

_El uso del término self para denominar al parámetro es solo un convenio en la jerga Pythonista para expresar que nos estamos refiriendo al propio (self) objeto, lo que realmente importa es que el primer parámetro que se le pasa al init será considerado como tal_.

```python
from math import sqrt

class Punto:
    def __init__(self, X, Y) -> None:
        self.x = X
        self.y = Y
        #esta función actua como una suerte de constructor e inicializador de instyancias de la clase Punto
        #para que ud no se confunda le hemos llamado X y Y (mayúscula) a los parámetros pero realmente no hay confunsión si se usan minúscula
        #porque no hay ambiguedad si se escribe self.x = x

    def distancia(punto1, punto2):
        return sqrt((punto1.x-punto2.x)**2 + (punto1.y-punto2.y)**2)

p1 = Punto(0,0)
p2 = Punto(3,4)
print(f"Distancia entre {p1} y {p2}: {Punto.distancia(p1,p2)}")
```

### 1.2.2 Funciones de la clase y uso del self

Las funciones que se definen dentro de la clase se pueden invocar con la sintaxis

_nombre_de_la_clase.nombre_de_la_funcion(parámetros)_

Esto es lo que se hizo en el código anterior con la función `distancia` para hacer la llamada con la sintaxis

`Punto.distancia(p1,p2)`

Sin embargo, siendo `c `una instancia de una clase `C` y `f` una función definida dentro de `C`, la sintaxis de Python permite hacer una llamada a la función `f `con la sintaxis

`c.f(parametros)`

la cual Python interpretará como una llamada de la forma `C.f(c, parametros)`

En la jerga de la POO `f` se denomina entonces un **método (función) de instancia**

Podemos tener entonces una nueva versión de la clase `Punto` con el método `distancia` definido como se muestra en el código abajo. De modo que cuando se hace entonces

`p1.distancia(p2)` puede leerse como se está llamando al método `distancia` de `p1` para que calcule su distancia a un punto` p2`

```python
from math import sqrt

class Punto:
    def __init__(self, x: int, y: int) -> None:
        self.x: int = x
        self.y: int = y

    def distancia(self, otro_punto: 'Punto') -> float:
        dx: float = otro_punto.x - self.x
        dy: float = otro_punto.y - self.y
        return sqrt(dx**2 + dy**2)

p1 = Punto(0,0)
p2 = Punto(3,4)
print(f"Distancia entre {p1} y {p2}: {p1.distancia(p2)}") #Se aplica el __str__ si está definido
```

### 1.2.3 Función especial __str__
Si algo aún no nos gusta del código anterior es que cuando se hace

`print(f"Distancia entre {p1} y {p2}` ....

lo que tenemos como salida de la impresión de ambos puntos es la información

`<__main__.Punto object at 0x0000020CEA5634D0>` y `<__main__.Punto object at 0x0000020CEA531D10>`

Lo que ha hecho Python al "visualizarnos" el valor de `p1` es decirnos que es un valor de un tipo Punto y darnos un código (que podemos interpretar con la posición en memoria donde está ubicada la información).

Cuando una variable tiene como valor un tuplo de dos enteros, como ocurre cuando se ejecuta el código

`p = (3,4)
print(f" p es {p}")`

Python sabe que el valor es un tuplo y que sus elementos son dos enteros 3 y 4 por tanto puede escribir

`p es (3,4)`

pero cuando `p1` es un objeto instancia de `Punto` como `Punto(3,4)`, objeto que tiene dos variables `x `y `y`, Python no tiene por qué "adivinar" cómo queremos que se visualice este valor.

Sin embargo, Python tiene otra función especial `__str__` que nos permite expresar cómo queremos representar como string un valor instancia de `Punto` y por consiguiente sea usado en una función como `print` que espera un string. Esto es lo que se hace en el código de más abajo. La función regresa el valor string

 `f"({self.x}, {self.y})"`

que es la cadena que comienza con el paréntesis abierto `"("` seguida por el valor de `self.x` (que es el string correspondiente valor entero) seguida por una coma `","` luego por el string correspondiente al valor entero `self.y` y termina con paréntesis abierto `")"`. De este modo cuando se ejecuta el código

`print(f"Distancia entre {p1} y {p2}: {p1.distancia(p2)}")`

Python escribirá el texto

`Distancia entre (0, 0) y (3, 4): 5.0`


```python
from math import sqrt
#versión de la clase Punto con
class Punto:
    def __init__(self, x: int, y: int):
        self.x: int = x
        self.y: int = y

    def distancia(self, otro_punto):
        dx: float = otro_punto.x - self.x
        dy: float = otro_punto.y - self.y
        return sqrt(dx**2 + dy**2)

    #Función especial que indica como queremos que se escriba el objeto Punto en forma de string
    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Punto(0,0)
p2 = Punto(3,4)
print(f"Distancia entre {p1} y {p2}: {p1.distancia(p2)}") #Se aplica el __str__ si está definido
```

Una conclusión muy simplista sería pretender que si una instancia c de una clase C tiene definidas unas variables v1, v2, ... vn entonces que Python asumiese que el __str__ de C fuese el string de un tuplo formado por los valores expresados como string de cada una de sus variables, algo así como (valor de v1 como string, valor de v2 como string, ....).

Aunque esta la situación de un ejemplo tan sencillo como el de la clase Punto, no puede generalizarse para clases que expresen objetos más complejos como iremos viendo a lo largo del curso. Por ello esto queda a cargo de lo que diseñe y codifique el programador.
