```python
#Reusando una clase en otra clase
from typing import List
from math import sqrt

class Punto:
    def __init__(self, x: int, y: int) -> None:
        self.x: int = x
        self.y: int = y

    def __str__(self) -> str:
        return f"({self.x}, {self.y})"

    def distancia(self, otro_punto: 'Punto') -> float:
        dx: float = otro_punto.x - self.x
        dy: float = otro_punto.y - self.y
        return sqrt(dx**2 + dy**2)

class Poligono:
    def __init__(self, vertices: List['Punto']) -> None:
        self.vertices: List['Punto'] = vertices

    def perimetro(self) -> float:
        total: float = 0.0
        n: int = len(self.vertices)

        for i in range(n):
            p1: Punto = self.vertices[i]
            p2: Punto = self.vertices[(i + 1) % n]
            total += p1.distancia(p2)

        return total

# Ejemplo de uso
p1: Punto = Punto(0, 0)
p2: Punto = Punto(3, 0)
p3: Punto = Punto(3, 4)
p4: Punto = Punto(0, 4)

poligono: Poligono = Poligono([p1, p2, p3, p4])
print(f"Poligono: {poligono} Perímetro: {poligono.perimetro()}")  # 16.0
```

## 2.1 Tipado dinámico

Mediante el recurso class Python permite encapsular y agrupar sintácticamente en una misma definición variables y funciones que trabajan sobre esas variables Python, con esto se está definiendo una suerte de **tipo de objetos** que tendrán una representación y una funcionalidad.

En Python todo puede considerarse un objeto y cada objeto es de un determinado tipo que indica cómo se representan sus valores y las operaciones que se pueden hacer con esos objetos. Están los tipos primitivos básicos `int, float, string, bool` que tienen una representación para sus valores y operaciones o funciones que se pueden hacer con ellos y que ya vienen integradas en Python. Estos son los valores básicos sobre los cuales se montará el andamiaje para definir otros objetos y tipos que formarán parte de un programa Python. Así tenemos también integrados en Python los tipos listas `[...]`, tuplos `(...)`, diccionarios `{llave:valor, llave:valor, ...}` formados a partir de ellos mismos o de cualquier tipo Python.

Con este recurso **class** Python nos permite crear nuevos tipos, darle sintáctica y semánticamente, definir cómo se forman sus objetos a partir de otros tipos ya definidos, y definir funciones (llamadas _métodos_ en la jerga de la POO) que trabajan o actuan sobre los objetos de dicho tipo.

La clase `Punto` a continuación nos permite crear puntos por sus coordenadas enteras y define una función `distancia` para calcular la distancia entre dos puntos.

La legibilidad y funcionalidad correcta de este código depende del estilo del programador a la hora de dar nombres y de aplicar las operaciones válidas según del tipo de los valores. Aunque se han utilizado nombres legibles, el código de más abajo no dice que los parámetros que se le pasea a x y y deben ser valores enteros, que el parámetro que se le pase a otro_punto debe ser a su vez del tipo Punto y que la función distancia devuelve un float.

Python controlará en ejecución que no se intenten aplicar operaciones inadecuadas sobre valores que no se corresponden con el tipo. Esto es lo que se denomina **tipado dinámico**. Pero no nos previene anticipadamente de cometer el error. Descomente los segmentos de código 1 y 2 del código de más abajo e intente ejecutarlo para que observe que ocurre.

El segmento 1 provocará un error cuando se intenta ejecutar `otro_punto.y - self.y `porque `otro_punto.y` sería un valor `str` y `self.y` es un valor `int` y no se puede aplicar una operación restar (-) entre un `str` y un `int`

El segmento 2 provocará error cuando al ser invocado el cálculo de `origen.distancia("mi_punto")` se intente ejecutar `dx = otro_punto.x - self.x` cuando realmente el valor que se le ha pasado al parámetro `otropunto` es el str `"mi_punto"` que por supuesto no tiene ninguna variable `x`
.

```python
from math import sqrt

class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distancia(self, otro_punto):
        dx = otro_punto.x - self.x
        dy = otro_punto.y - self.y
        return sqrt(dx**2 + dy**2)

origen = Punto(0,0)
p = Punto(3,4)
print(f"Distancia entre {origen} y {p}: {origen.distancia(p)}")

#DESCOMENTE Y PRUEBE LOS CÓDIGOS A CONTINUACIÓN PARA QUE APRECIE LOS ERROES
#ERROR 1 Parámetro inadecuado al intentar usar como int una variable que es str "abc"
# p2 = Punto(5, "abc")
# print(f"Distancia entre {origen} y {p2}: {origen.distancia(p2)}")

# ERROR 2 Parámetro inadecuado al intentar pasar un str como parámetro de tipo Punto a la función distancia que lo que espera es un punto
print(f"Distancia entre {origen} y {"mi_punto"}: {origen.distancia("mi_punto")}")
```

## 2.2 Poniendo tipos sintácticamente en Python
Python tiene un recurso que se denomina **Python anotado** que permite usar explicitamente los nombres de tipos en la sintaxis de Python para "anotar" de que tipo es una varible, un parámetro, el valor que retorna una función, los elementos de una lista, la llave y el valor asociado en un diccionaio .... El código anterior de la clase `Punto` quedaría más legible y mejor definido si escribiera en la forma siguiente:

```python
from math import sqrt

class Punto:
    def __init__(self, x: int, y: int) -> None:
        self.x: int = x
        self.y: int = y

    def distancia(self, otro_punto: Punto) -> float:
        dx: float = otro_punto.x - self.x
        dy: float = otro_punto.y - self.y
        return sqrt(dx**2 + dy**2)

    def __str__(self) -> str:
        return f"({self.x}, {self.y})"

origen = Punto(0,0)
p = Punto(3,4)
print(f"Distancia entre {origen} y {p}: {origen.distancia(p)}") #Se aplica el __str__ si está definido
```

Al escribir `def __init__(self, x: int, y: int) -> None:` estamos diciendo que **los parámetros** `x` y `y` **son de tipo** `int` y que la función especial `__int__` **devuelve** `None` (es decir no devuelve directamente ningún valor)

`self.x: int = x` indica que **la variable** `self.x` que se está definiendo para **es de tipo** `int`

`def distancia(self, otro_punto: Punto) -> float:` indica que **el parámetro** `otro_punto` **debe ser de tipo** `Punto` y que **la función devuelve un** `float`

`dx:float = ...` nos dice que **la variable** `dx` **es de tipo** `float`

`def __str__(self) -> str:` **nos indica que la función** `__str__` **devuelve un** `str`



## 2.3 Clases que usan otras clases
En la definición de una clase se pueden usar otras clases.

La siguiente definición de una clase `Poligono` importa el tipo `List` del módulo `typing`. `List[T]` expresa a un tipo que es una lista `list` pero que todos son elementos son de tipo `T`, de modo que `List[Punto]` denota a un tipo que es una _lista de puntos_.

`def __init__(self, vertices: List[Punto])` estamos diciendo que el parámetro `vertices` para construir un `Poligono` debe ser una _lista de puntos_, lo cual es el caso cuando se hace

`p1: Punto = Punto(0, 0)
p2: Punto = Punto(3, 0)
p3: Punto = Punto(3, 4)
p4: Punto = Punto(0, 4)
poligono: Poligono = Poligono([p1, p2, p3, p4]) `

Esto nos aporta **más claridad y legibilidad** en la definición y uso del código `Poligono` aunque, por la naturaleza de tipado dinámico, no nos previene de cometer el error de colocar algo que no sea un `Punto` en el supuesto parámetro que debiera ser una lista de puntos. Descomente en el código la secuencia ERROR que intenta colocar un parámetro incorrecto al hacer

`poligono: Poligono = Poligono([p1, p2, p3, "soy un string no un punto"])`
`print("\nHasta aquí todo va bien!!...")`

`print(f"Poligono: {poligono} Perímetro: {poligono.perimetro()}")`

Python no nos da excepcion al hacer

`poligono: Poligono = Poligono([p1, p2, p3, "soy un string no un punto"])` porque realmente intenta (por su naturaleza de tipado dinámico) ejecutar el código de esa asignación, que en sí misma no es problema, pero que ha colocado como supuesto vértice algo que no es un `Punto`. La excepción la dará cuando se intenta ejecutar el código

`poligono.perimetro()`

y se manifestará en el oscuro mensaje

**AttributeError: 'str' object has no attribute 'x'** porque en alguna parte le pedirá por el valor de `x` a un valor que es realmente el string `"soy un string no un punto" `que por supuesto no tiene ninguna variable `x`

_Realmente hay maneras de lograr que Python nos de un error más elocuente y sin llegar a ejecutar el código, pero esto lo veremos más adelante. Lo que nos interesa por ahora es que usted use este concepto de tipado para escribir un código más claro y legible y a la vez comprender de la misma manera el código escrito por otros_



```python
#Reusando una clase en otra clase
from typing import List
from math import sqrt

class Punto:
    def __init__(self, x: int, y: int) -> None:
        self.x: int = x
        self.y: int = y

    def __str__(self) -> str:
        return f"({self.x}, {self.y})"

    def distancia(self, otro_punto: Punto) -> float:
        dx: float = otro_punto.x - self.x
        dy: float = otro_punto.y - self.y
        return sqrt(dx**2 + dy**2)

class Poligono:
    def __init__(self, vertices: List[Punto]) -> None:
        self.vertices:List[Punto] = vertices
        if len(self.vertices) < 3:
            raise ValueError(
              f"Un polígono necesita al menos 3 puntos, recibido: {len(self.vertices)}"
            )

    def __str__(self) -> str:
        return "[" + ",".join(str(v) for v in self.vertices) + "]"

    def perimetro(self) -> float:
        total: float = 0.0
        n: int = len(self.vertices)

        for i in range(n):
            p1: Punto = self.vertices[i]
            p2: Punto = self.vertices[(i + 1) % n]
            total += p1.distancia(p2)
        return total

# Ejemplo de uso
p1: Punto = Punto(0, 0)
p2: Punto = Punto(3, 0)
p3: Punto = Punto(3, 4)
p4: Punto = Punto(0, 4)

poligono: Poligono = Poligono([p1, p2, p3, p4])
print(f"Poligono: {poligono} Perímetro: {poligono.perimetro()}")  # 16.0

poligono: Poligono = Poligono([p1, p2, p3, "soy un string no un punto"])
print("\nHasta aquí todo va bien!!...")
print(f"Poligono: {poligono} Perímetro: {poligono.perimetro()}")  # 16.0
```

### 2.5.1 El decorador dataclass
Puede parecer repetitivo y cansón tener que escribir explícitamente, en el código de `__init__`, la asignación de cada parámetro del __init__ a la correspondiente variable de instancia:

```python
class Film:
    def __init__(
        self,
        titulo: str,
        pais: list[str],
        fecha: int,
        director: str,
        generos: list[str],
        actores: list[str],
        rating_imdb: float,
    ) -> None:
        self.titulo = titulo
        self.pais = pais
        self.fecha = fecha
        self.director = director
        self.generos = generos
        self.actores = actores
        self.rating_imdb = rating_imdb
```

Esto se puede escribir de modo más simple usando el recurso de `dataclass:`

```python
import json
from typing import List
from dataclasses import dataclass

@dataclass
class Film:
    titulo: str
    pais: List[str]
    fecha: int           # año de estreno
    director: str
    generos: List[str]
    actores: List[str]
    rating_imdb: float
```

En este caso Python nos genera automáticamente el código del `__init__` con la inicialización de las variables a partir de los parámetros. Este `@dataclass` es lo que se denomina un **decorador**, en este caso debe ponerse delante de la definición de la clase y con el símbolo **@**. Este recurso de decorador será estudiado con más detenimiento más adelante donde veremos cómo definir y aplicar decoradores. Por ahora sugerimos usarlo en la definición de una clase para simplificar y hacer más legible y robusto el código

## 2.6 De un json a una lista de objetos

Si antes habíamos visto como cargar una archivo textual en formato json a una lista de diccionarios. Ahora podemos hacerlo hacia una lista de objetos en Python, en este ejemplo a una lista de objetos Film. Procesar entonces la información como objetos de un cierto tipo nos aporta más legibilidad, protección y claridad.

En el siguiente código la función `json.load(f)` nos carga el archivo json y lo representa como una lista de diccionarios. Luego al hacer

`[Film(**d) for d in films_dicc]`

se recorre la lista de diccionarios. El efecto del `**d` es que cada diccionario es desempaquetado en la forma de la lista de parámetros que se le pasa al constructor `__init__` de `Film`. Por supuesto que para que esto funcione correctamente los valores usados con llaves en los diccionarios json tienen que coincidir con los nombres de los parámetros del `__init__` (que son a su vez los nombres de las varables de instancia.

El código siguiente nos listaría entonces los títulos de los films guardados en el archivo json


```python
import json
from typing import List
from dataclasses import dataclass

@dataclass
class Film:
    titulo: str
    pais: List[str]
    fecha: int
    director: str
    generos: List[str]
    actores: List[str]
    rating_imdb: float

def cargar_films_desde_json(ruta_archivo_json: str) -> List[Film]:
    #Funcion que lee un archivo json que represente una lista de diccionarios que correspondan a un film y devuelve una lista de objetos Film
    with open(ruta_archivo_json, 'r', encoding='utf-8') as f:
      films_dicc = json.load(f) #Carga una lista de diccionarios del archivo json

    return [Film(**d) for d in films_dicc]
    #el operador **d desempaqueta un diccionario del estilo {"titulo":"El Padrino I", "pais":"Estados Unidos" ....} en código como
    #Film(titulo="El Padrino 1", pais="Estados Unidos",...)

archivo_json = input("Escriba ruta del archivo json: ") # MK_films.json ubicado en la misma carpeta del proyecto
mis_films_favoritos = cargar_films_desde_json(archivo_json)
for f in mis_films_favoritos: print(f.titulo)
```

## 2.7 Variables de clase

Una variable de clase es una variable definida dentro de la clase pero no asociada a cada instancia sino a toda la clase. Esta variable podría ser accedida desde el código de cualquier función dentro de la clase.

Supongamos para nuestro ejemplo de films que queremos dar automáticamente un número a cada film de la lista. El código que sigue nos muestra cómo definir dentro de la clase `Film` una varible `total_films` inicializada en 0 para que sea incrementada cada vez que se crea una instancia. La especificación `ClassVar[int]` nos indica que es una **variable de clase** cuyo valor senespera sea un `int` y que en este caso además incializa en `0`

 `total_films: ClassVar[int] = 0`

Pero si el código de `__init__` fue generado automáticamente al decorar la clase con `@dataclass` ¿dónde  incrementar `total_films` cada vez que se crea una instancia?

En Python se puede definir también otro método especial __post_init__ que ejecutaría despues del __init__
llevar la cuenta de cuántas instancias de `Film` se van creando. Lo que se hace en el código es añadirle una variable de instancia, le hemos llamado `id_film` y actualizar esa variable con el valor `total_films` cada vez que se cree un Film.

En Python se puede definir una función `__post_init__` que se aplicaría automáticamente después de `__init__`. En el código de esta función es que se crearía entonces la variable de instancia `id_film` que a la que se le asignaría el valor de `total_films` como se puede apreciar en el código siguiente.

```python
import json
from typing import List, ClassVar
from dataclasses import dataclass

@dataclass
class Film:
    titulo: str
    pais: List[str]
    fecha: int
    director: str
    generos: List[str]
    actores: List[str]
    rating_imdb: float
    total_films: ClassVar[int] = 0

    def __post_init__(self):
        Film.total_films = Film.total_films + 1
        self.id_film = self.total_films

def cargar_films_desde_json(ruta_archivo_json: str) -> List[Film]:
    #Funcion que lee un archivo json que represente una lista de diccionarios que correspondan a un film y devuelve una lista de objetos Film
    with open(ruta_archivo_json, 'r', encoding='utf-8') as f:
      films_dicc = json.load(f) #Carga una lista de diccionarios del archivo json

    return [Film(**d) for d in films_dicc]
    #el operador **d desempaqueta un diccionario del estilo {"titulo":"El Padrino I", "pais":"Estados Unidos" ....} en código como
    #Film(titulo="El Padrino 1", pais="Estados Unidos",...)

archivo_json = input("Escriba ruta del archivo json: ") # MK_films ubicado en la misma carpeta del proyecto
mis_films_favoritos = cargar_films_desde_json(archivo_json)
for f in mis_films_favoritos: print(f"{f.id_film:4}  {f.titulo}")
```
