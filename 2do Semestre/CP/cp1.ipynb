1. Defina una clase `Fecha` para expresar fechas que tengan día, mes y año y operaciones asociadas a este concepto de Fecha

```python
class Fecha: 
    def __init__(self, day, month, year):
        self.d: int = day
        self.m: int = month
        self.y: int = year

    def __str__(self):
        return f"{self.d}/{self.m}/{self.y}"
    
hoy = Fecha(3, 11, 2005)
print(hoy)
```
