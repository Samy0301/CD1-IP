🧩 Resolver Sudoku [Leetcode Link](https://leetcode.com/problems/sudoku-solver/description/)

Enunciado:
Escribe un programa que resuelva un rompecabezas de Sudoku llenando las celdas vacías.
Una solución de Sudoku debe cumplir todas las siguientes reglas:
- Cada dígito del 1 al 9 debe aparecer exactamente una vez en cada fila.
- Cada dígito del 1 al 9 debe aparecer exactamente una vez en cada columna.
- Cada dígito del 1 al 9 debe aparecer exactamente una vez en cada uno de los 9 subcuadros de 3x3 de la cuadrícula.
El carácter '.' indica una celda vacía.

📖 Ejemplo
Entrada:
```
board = [
["5","3",".",".","7",".",".",".","."],
["6",".",".","1","9","5",".",".","."],
[".","9","8",".",".",".",".","6","."],
["8",".",".",".","6",".",".",".","3"],
["4",".",".","8",".","3",".",".","1"],
["7",".",".",".","2",".",".",".","6"],
[".","6",".",".",".",".","2","8","."],
[".",".",".","4","1","9",".",".","5"],
[".",".",".",".","8",".",".","7","9"]
]
```

![Problema](img/cp13_sudoku.png)

Salida:
```
[
["5","3","4","6","7","8","9","1","2"],
["6","7","2","1","9","5","3","4","8"],
["1","9","8","3","4","2","5","6","7"],
["8","5","9","7","6","1","4","2","3"],
["4","2","6","8","5","3","7","9","1"],
["7","1","3","9","2","4","8","5","6"],
["9","6","1","5","3","7","2","8","4"],
["2","8","7","4","1","9","6","3","5"],
["3","4","5","2","8","6","1","7","9"]
]
```

![Solución](img/cp13_sudoku_solution.png)

🔒 Restricciones
- board.length == 9
- board[i].length == 9
- board[i][j] es un dígito '1'–'9' o '.'.
- Se garantiza que el tablero de entrada tiene una única solución.

````python
def solveSudoku( board: List[List[str]]) -> None:
    """
    Do not return anything, modify board in-place instead.
    """
    for i in range(9):
        for j in range(9):
            if board[i][j] == ".":
                for n in range(1, 10):
                    if es_valido(board, i, j, n):
                        board[i][j] = f"{n}"
                        if solveSudoku(board):
                            return True
                        board[i][j] = "."
                return False
    return True
    

def es_valido(tab, f, c, n):
    # fila
    for j in range(9):
        if tab[f][j] == f"{n}": return False
    # columna
    for i in range(9):
        if tab[i][c] == f"{n}": return False
    # bloque 3×3
    bf, bc = (f // 3) * 3, (c // 3) * 3
    for i in range(3):
        for j in range(3):
            if tab[bf + i][bc + j] == f"{n}": return False
    return True

def imprimir(tab):
    for fila in tab:
        print("".join(str(x) for x in fila))

board = [
["5","3",".",".","7",".",".",".","."],
["6",".",".","1","9","5",".",".","."],
[".","9","8",".",".",".",".","6","."],
["8",".",".",".","6",".",".",".","3"],
["4",".",".","8",".","3",".",".","1"],
["7",".",".",".","2",".",".",".","6"],
[".","6",".",".",".",".","2","8","."],
[".",".",".","4","1","9",".",".","5"],
[".",".",".",".","8",".",".","7","9"]
]

if solveSudoku(board): 
    imprimir(board)
````
