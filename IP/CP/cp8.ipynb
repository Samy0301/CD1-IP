3 Escriba una función que devuelva un string conteniendo solo las consonantes del string que recibe como argumento. Puede asumir que el string recibido solo tiene minúsculas.
```
Ejemplo
print(keep_consonants("abcd"))  # prints bcd
print(keep_consonants("aaa"))  # prints an empty string
print(keep_consonants("babas"))  # prints bbs
```

````python
def keep_consonants(word):
    """ word is a string of lowercase letters
        Returns a string containing only the consonants
        of word in the order they appear
    """
    vocales = ["a", "e", "i", "o" , "u"]
    w = ""
    for i in word:
        if i not in vocales:
            w += i
    return w

print(keep_consonants("abcd"))
print(keep_consonants("aaa"))
print(keep_consonants("babas"))
````

4 Escriba una función que reciba un string y un carácter. Dicha función debe calcular la diferencia entre los índices de la primera y la última vez que ocurre ese carácter en el string. Si el caracter no está entonces se debe imprimir -1

```
Ejemplo
print(first_to_last_diff('aaaa', 'a'))  # prints 3
print(first_to_last_diff('abcabcabc', 'b'))  # prints 6
print(first_to_last_diff('xyz', 'b'))  # prints -1
```

````python
def first_to_last_diff(s, c):
    """ s is a string, c is single character string
        Returns the difference between the index where c first
        occurs and the index where c last occurs. If c does not
        occur in s, returns -1.
    """
    if c not in s: return -1

    first = 0
    last = 0
    for i in range(len(s)):
        if s[i] == c:
            first = i+1
            break
    for i in range(len(s)):
        if s[i] == c: 
            last = i+1
    
    return last - first

print(first_to_last_diff("aaaa", "a"))
print(first_to_last_diff("abcabcabc", "b"))
print(first_to_last_diff("xyz", "b"))
````

5 Escriba una función que reciba una cantidad incierta de valores decimales y devuelva el promedio de todos ellos

```
Ejemplo
print(prom(5, 5, 5, 5, 6, 6, 6, 6))  # prints 5.5
print(prom(1.0, 3.0)  # prints 2
```

````python
def prom(*args):
    sum = 0
    for i in args:
        sum += i
    return sum/len(args)

print(prom(5,5,5,5,6,6,6,6))
print(prom(1.0, 3.0))
````

8 Escriba una función que reciba como argumentos las entradas que deben ser añadidos a un diccionario global
```
Ejemplo
add2dic(a = 86)
print(dict) # prints { "a": 86 }
add2dic(b = 63)
print(dict) # prints { "a": 86, "b": 63 }
add2dic(c = 1, d = 2, e = 3)
print(dict) # prints { "a": 86, "b": 63, "c": 1, "d": 2, "e": 3 }
```

````python
def add2dic(dic = {}, **kwargs):
    for key, value in kwargs.items():
        dic.update({key : value})
    return dic

print(add2dic(a = 86))
print(add2dic(b = 63))
add2dic(h = 7)
print(add2dic(c = 1, d = 2, e = 3))
````
