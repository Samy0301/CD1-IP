```python
import math
import random

# TRES CICLOS ANIDADOS. SOLUNCION DE ORDEN n^3
def sub_maxsum_fuerzabruta(list):
    """
    Encuentra la subsecuencia contigua con máxima suma por FUERZA BRUTA.
    Prueba TODAS las subsecuencias posibles con 3 bucles anidados (O(n³)).

    Args:
        list (list): Lista de enteros (puede incluir negativos)

    Returns:
        tuple: (suma_maxima, indice_inicio, indice_fin)
    """
    if len(list) == 0:
        return 0, -1, -1 #Caso lista vacía devuelve el triplo (0, -1, -1)
    
    suma_maxima = list[0] #Asumimos que el primer elemento de la lista es en sí mismo la subsecuencia de máxima suma
    inicio_max, fin_max = 0, 0
    
    # Ciclo 1: inicio de subsecuencia
    for i in range(len(list)):
        # Ciclo 2: fin de subsecuencia
        for j in range(i, len(list)):
            # Ciclo 3: sumar la subsecuencia de la posición i hasta j
            suma_actual = 0
            for k in range(i, j + 1):
                suma_actual += list[k]
            # Quedarnos con esta suma si es mayor que la que teníamos y actualizar el inicio y fin
            if suma_actual > suma_maxima:
                suma_maxima = suma_actual
                inicio_max = i
                fin_max = j
    
    return suma_maxima, inicio_max, fin_max

# Pruebas
print(sub_maxsum_fuerzabruta([2, -1, -3, 5, -2, -1, 6, 4, -8])) # (12, 3, 7)
print(sub_maxsum_fuerzabruta([1, -2, 3, 4, -5, 6, -1, 2]))  # (9, 2, 7)
print(sub_maxsum_fuerzabruta([-2, -3, 4, -1, -2, 1, 5, -3]))  # (7, 2, 6)
print(sub_maxsum_fuerzabruta([-1, -2, -3]))  # (-1, 0, 0)
print(sub_maxsum_fuerzabruta([5]))  # (5, 0, 0)
```

```python
#DOS CICLOS ANIDADOS. SOLUNCION DE ORDEN n^2
def sub_maxsum_cuadratica(list):
    if len(list)==0:
        return 0, -1, -1

    suma_maxima = list[0]
    inicio_max, fin_max = 0, 0

    # Ciclo 1: inicio de subsecuencia
    for i in range(len(list)):
        suma_actual = 0
        # Ciclo 2: fin de subsecuencia (suma acumulativa desde i)
        for j in range(i, len(list)):
            # A la vez que se va recorriendo el intervalo se va haciendo la suma para irnos quedando con la máxima
            suma_actual += list[j]
            if suma_actual > suma_maxima:
                suma_maxima = suma_actual
                inicio_max = i
                fin_max = j

    return suma_maxima, inicio_max, fin_max

# Pruebas
print(sub_maxsum_cuadratica([10, -1, -3, 5, -2, -1, 6, 4, -8])) # (12, 3, 7)
print(sub_maxsum_cuadratica([1, -2, 3, 4, -5, 6, -1, 2]))  # (9, 2, 7)
print(sub_maxsum_cuadratica([-2, -3, 4, -1, -2, 1, 5, -3]))  # (7, 2, 6)
print(sub_maxsum_cuadratica([-3, -10, -3]))  # (-1, 0, 0)
print(sub_maxsum_cuadratica([5]))  # (5, 0, 0)
print(sub_maxsum_cuadratica([-10,0,-3]))  # (5, 0, 0)
```

```python
#CON DIVIDE Y VENCERAS HACIENDO RECURSION ORDEN log2(n)*n
def subSumMaxDivVenc(lista):
    def subSumMaxRec(lista, inf, sup):
        maxHaciaIzq = maxHaciaDer = 0
        if (inf == sup):
            # El intervalo es de un solo elemento su sub de max suma es ese elemento
            return (lista[inf], inf, inf) # if lista[inf] > 0 else 0
        else:
            medio = (inf+sup)//2
            #Calcular máxima subsecuencia que empieza y termina en la mitad izquierda
            maxIzq, izqInicio, izqFin = subSumMaxRec(lista, inf, medio)
            #Calcular máxima sunsecuencia que empieza y termina en la mitad derecha
            maxDer, derInicio, derFin = subSumMaxRec(lista, medio+1, sup)
            #Calcular máxima subsecuencia que incluye de la mitad de la izquierda y de la derecha
            maxHaciaIzq = None
            medioInicio = None
            sum = 0
            # Recorrer desde el medio hacia la izquierda
            for i in range(medio, -1, -1):
                sum += lista[i]
                if maxHaciaIzq is None or sum > maxHaciaIzq:
                    maxHaciaIzq = sum
                    medioInicio = i

            maxmedioHaciaDer = None
            medioFin = None
            sum = 0
            # Recorrer desde el medio hacia la izquierda
            for i in range(medio+1, sup+1):
                sum += lista[i]
                if maxHaciaDer is None or sum > maxHaciaDer:
                    maxHaciaDer = sum
                    medioFin = i

            if maxHaciaIzq == None:
                maxMedio = maxHaciaDer
            elif maxHaciaDer == None:
                maxMedio = maxHaciaIzq
            else: maxMedio = maxHaciaIzq+maxHaciaDer
            # elif maxHaciaIzq >= maxHaciaDer:
            #     maxreturn (maxHaciaIzq, medioInicio, medio)
            # else:
            #   return (maxHaciaDer, medio+1, medioFin)
            #           maxMedio, medioInicio, medioFin = (maxHaciaIzq + maxHaciaDer, i1, j1)
            if maxIzq >= max(maxMedio, maxDer): return(maxIzq, izqInicio, izqFin)
            elif maxMedio >= max(maxIzq, maxDer): return (maxMedio, medioInicio, medioFin)
            else: return (maxDer, derInicio, derFin)

    if len(lista) == 0: #lista vacía no tiene secuencia de suma max
        result = (0, -1, -1)
    else:
        result = subSumMaxRec(lista, 0, len(lista)-1)
    return result

lista = [2, -1, -3, 5, -2, -1, 6, 4, -8, 1] #(12, 3, 7)
print(subSumMaxDivVenc(lista))
lista = [12, -1, -3, 5, -13, -1, 6, 4, -8, 1] #(13, 0, 3)
print(subSumMaxDivVenc(lista))
lista = [2, -1, -3, 5, -12, -1, 6, 4, -8, 1] #(10, 6, 7)
print(subSumMaxDivVenc(lista))
lista = [] #(0, -1, -1)
print(subSumMaxDivVenc(lista))
lista = [-10, 0, -3] #(0, 1, 1)
print(subSumMaxDivVenc(lista))
lista = [-10, -2, -3] #(-2, 1, 1)
print(subSumMaxDivVenc(lista))
lista = [0] #(0, 0, 0)
print(subSumMaxDivVenc(lista))
lista = [100, -1000] #(100, 0, 0)
print(subSumMaxDivVenc(lista))
lista = [-1000, 100] #(100, 1, 1)
print(subSumMaxDivVenc(lista))
lista = [2, -10, -10, 4, -20, -20, 3]
print(subSumMaxDivVenc(lista)) #(4, 3, 3)
```

```python
# SOLUCION DE UN SOLO CICLO, DE ORDEN LINEAL n
def subSumMaxLineal(lista):
    """
    Devuelve (max_suma, inicio, fin) de la subsecuencia contigua
    con suma máxima en arr. Complejidad O(n).
    """
    if not lista:
        return (0, -1, -1)

    suma_actual = lista[0]      # mejor suma terminando en la posición actual
    suma_maxima = lista[0]      # mejor suma encontrada hasta ahora

    inicio_actual = 0           # inicio de la subsecuencia actual
    mejor_inicio = 0            # mejor inicio encontrado
    mejor_fin = 0               # mejor fin encontrado

    for i in range(1, len(lista)):
        # Decidir si es mejor empezar una nueva subsecuencia en i
        # o continuar la subsecuencia anterior
        if lista[i] > suma_actual + lista[i]:
            suma_actual = lista[i]
            inicio_actual = i
        else:
            suma_actual += lista[i]

        # Actualizar la mejor subsecuencia global si mejora
        if suma_actual > suma_maxima: #si encontramos una secuencia que suma más que la que mayor sumaba hasta ahora nos quedamos con esa nueva
            suma_maxima = suma_actual
            mejor_inicio = inicio_actual
            mejor_fin = i

    return suma_maxima, mejor_inicio, mejor_fin

lista = [2, -1, -3, 5, -2, -1, 6, 4, -8, 1] #(12, 3, 7)
print(subSumMaxLineal(lista))
lista = [12, -1, -3, 5, -13, -1, 6, 4, -8, 1] #(13, 0, 3)
print(subSumMaxLineal(lista))
lista = [2, -1, -3, 5, -12, -1, 6, 4, -8, 1] #(10, 6, 7)
print(subSumMaxLineal(lista))
lista = [] #(0, -1, -1)
print(subSumMaxLineal(lista))
lista = [-10, 0, -3] #(0, 1, 1)
print(subSumMaxLineal(lista))
lista = [-10, -2, -3] #(-2, 1, 1)
print(subSumMaxLineal(lista))
lista = [0] #(0, 0, 0)
print(subSumMaxLineal(lista))
lista = [100, -1000] #(100, 0, 0)
print(subSumMaxLineal(lista))
lista = [-1000, 100] #(100, 1, 1)
print(subSumMaxLineal(lista))
lista = [2, -10, -10, 4, -20, -20, 3]
print(subSumMaxLineal(lista)) #(4, 3, 3)
```
