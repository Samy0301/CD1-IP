```python
# funcion que dado 2 listas devuelva la intercepcion
def intercepcion(l1, l2):
    result=[]
    for i in l1:
        if i in l2 and i not in result:
            result.append(i)
    return result

print(intercepcion([2,3,4,4], [2, 3,4,5]))
```

```python
# recibir 2 listas l1 valores y l2 mascara donde si mascara es true entonces el elemento de l1 va a la respuesta ambas con el mismo len
def eres_true(lst, mask):
    result=[]
    for i in range(len(lst)):
        if mask[i]: result.append(lst[i])
    return result

a=[1,2]
b=[True,False]
print(eres_true(a,b))
```
