## 4Ô∏è‚É£ List-comprehensions + filtros

## 7Ô∏è‚É£ Ordenar sin `sorted()` (selection-sort)

```python
# Ordenar cadenas por longitud (CP5-ex1)
lst = ["Mar√≠a", "Juan", "Adalberto"]
for i in range(len(lst)):
    m = i
    for j in range(i+1, len(lst)):
        if len(lst[j]) < len(lst[m]):
            m = j
    lst[i], lst[m] = lst[m], lst[i]
print("por longitud ‚Üí", lst)
```

## 9Ô∏è‚É£ Funciones vari√°dicas

```python
# Promedio de cualquier cantidad de n√∫meros (CP8-ex5)
def promedio(*args):
    return sum(args) / len(args) if args else 0

print(promedio(5, 5, 6, 6))   # 5.5

# Diccionario global con **kwargs (CP8-ex8)
GLOBAL_DIC = {}
def add2dic(**kw):
    GLOBAL_DIC.update(kw)
    return GLOBAL_DIC

add2dic(a=1, b=2)
print("global dic ‚Üí", GLOBAL_DIC)
```

## üîü Problema integrado (usa 8 trucos distintos)

**Enunciado r√°pido**:  
Lee enteros hasta l√≠nea vac√≠a ‚Üí quita repetidos ‚Üí ordena ‚Üí imprime los impares elevados al cubo, separados por " | ".

```python
import math, json

# 1) Entrada variable hasta ENTER
nums = list(map(int, iter(input, '')))   # truco iter(callable, sentinel)

# 2) Eliminar repetidos (set)
uniq = sorted(set(nums))                 # 3) sorted + set

# 4) List-comprehension con condici√≥n + potencia
cubos = [x**3 for x in uniq if x % 2]

# 5) Desempaquetado en print con separador personalizado
print(*cubos, sep=' | ')
```
