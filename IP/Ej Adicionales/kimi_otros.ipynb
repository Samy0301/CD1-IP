```python
def factoriza_primero_no_primo(n: int) -> list[int]:
    """
    Dado un entero n ≥ 2, devuelva la **factorización prima** del
    **primer número compuesto** que aparezca en la sucesión:
        n, n+1, n+2, ...
    La salida debe ser la lista de factores primos en orden creciente.

    Ejemplos
    --------
    >>> factoriza_primero_no_primo(7)   # 7 es primo, 8 es el primer compuesto
    [2, 2, 2]
    >>> factoriza_primero_no_primo(9)   # 9 ya es compuesto
    [3, 3]
    """
    while es_primo(n):
        n += 1

    factores = []
    # 1. Quitamos los 2
    aux = n
    while aux % 2 == 0:
        factores.append(2)
        aux = aux // 2
    # 2. Divisores impares
    d = 3
    lim = ((aux)**0.5) + 1
    while d <= lim:
        while aux % d == 0:
            factores.append(d)
            aux = aux // d
            lim = ((aux)**0.5) + 1
        d += 2
    # 3. Resto primo
    if aux > 1:
        factores.append(aux)

    return factores

def es_primo(n: int):
    if n <= 1: return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(factoriza_primero_no_primo(80))
```

```python
def count_up_runs(nums: list[int]) -> int:
    """
    Cuenta cuántas sub-secuencias **estrictamente crecientes**
    de longitud ≥ 2 existen dentro de `nums` (orden original).
    Una sub-secuencia no tiene por qué ser contigua.
    Ej.  [1,3,2,4]  ->  7  [(1, 3), (1, 2), (1, 4), (3, 4), (2, 4), (1, 3, 4), (1, 2, 4)]
    """
    
    n = 2
    lst_t = []
    while n <= len(nums):
        t = ()
        i = 0                       
        while True:
            if len(t) == n:
                lst_t.append(t)
                i = nums.index(t[-1]) + 1   # t[-1] = ultimo elemento en t, nums.index(x) = indice en el que x aparece en nums
                t = t[:-1]               # t[:-1] = quita el ultimo elemento en t     
                continue
            
            if i < len(nums):
                t += (nums[i],)
                i += 1
            else:
                if not t:
                    break
                i = nums.index(t[-1]) + 1
                t = t[:-1]
        n += 1
    
    lst=[]
    for tupla in lst_t:
        for i in range(len(tupla)-1):
            if tupla[i]<tupla[i+1]:
                if i+1 == len(tupla)-1:
                    lst.append(tupla)
            else:
                break
    
    return len(lst)


print(count_up_runs([1,3,2,4]))
```

```python
def merge_sorted_streams(*streams: list[int]) -> list[int]:
    """
    Reciba cualquier cantidad de listas **ya ordenadas asc.** y
    devuelva **una sola lista** con todos los elementos
    **en orden** (sin usar `sorted` ni `heapq`).
    Debe funcionar en O(total_elements).

    Ejemplo
    -------
    >>> merge_sorted_streams([1, 4], [], [2, 3, 7])
    [1, 2, 3, 4, 7]
    """
        
    # índices de lectura para cada lista
    idx = [0] * len(streams)
    merged = []

    while True:
        # búsqueda del mínimo entre las cabezas disponibles
        min_val = None
        min_pos = -1

        for i, cur in enumerate(idx):
            if cur < len(streams[i]):          # aún quedan elementos
                if min_val is None or streams[i][cur] < min_val:
                    min_val = streams[i][cur]
                    min_pos = i

        if min_pos == -1:                      # no quedan elementos
            break

        merged.append(min_val)
        idx[min_pos] += 1                      # avanzar cursor

    return merged

print(merge_sorted_streams([1, 4], [], [2, 3, 7]))
```

```python
def polinomio_desde_string(s: str) -> list[int]:
    """
    Convierta un string como "3x^3 -2x +7" en la lista de
    coeficientes enteros **completos** (incluyendo 0):
    [3, 0, -2, 7]  → 3x³ + 0x² -2x + 7
    Se garantiza:
        - solo coeficientes enteros
        - variable siempre 'x'
        - exponentes ≥ 0 y en orden decreciente
        - puede faltar término independiente o intermedio

    Ejemplo
    -------
    >>> polinomio_desde_string("x^2 - 1")
    [1, 0, -1]
    >>> polinomio_desde_string("-5x^3 +4x")
    [-5, 0, 4, 0]
    """
    term: dict[int, int] = {}
    i, n = 0, len(s)

    while i < n:
        # ---------- 1. signo ----------
        if s[i] == '+':
            sign, i = 1, i + 1
        elif s[i] == '-':
            sign, i = -1, i + 1
        else:
            sign = 1

        # ---------- 2. consumir espacios ----------
        while i < n and s[i] == ' ':
            i += 1

        # ---------- 3. coeficiente ----------
        coef_str = ''
        while i < n and s[i].isdigit():
            coef_str += s[i]
            i += 1
        coef = int(coef_str) * sign if coef_str else sign

        # ---------- 4. variable / exponente ----------
        if i < n and s[i] == 'x':
            i += 1
            # consumir posible '^'
            if i < n and s[i] == '^':
                i += 1
                exp_str = ''
                while i < n and s[i].isdigit():
                    exp_str += s[i]
                    i += 1
                exp = int(exp_str)
            else:
                exp = 1
        else:
            exp = 0

        term[exp] = term.get(exp, 0) + coef

        # ---------- 5. consumir espacios finales ----------
        while i < n and s[i] == ' ':
            i += 1

    if not term:
        return [0]

    grado_max = max(term)
    return [term.get(g, 0) for g in range(grado_max, -1, -1)]


print(polinomio_desde_string("x^2 - 1"))        # [1, 0, -1]
print(polinomio_desde_string("-5x^3 +4x"))      # [-5, 0, 4, 0]
print(polinomio_desde_string("3x^3 -2x +7"))    # [3, 0, -2, 7]
print(polinomio_desde_string("42"))             # [42]
print(polinomio_desde_string("-x^5 + x^2 - x")) # [-1, 0, 0, 1, -1, 0]
```

```python
def password_score(pwd: str) -> int:
    """
    Devuelva un **puntaje entero** según:
    +2 por cada clase de caracter presente (mayúsc, minús, dígito, símbolo)
    +1 por cada carácter único que no se repita
    -3 si aparece "123", "abc" o "qwe" (case-insensitive) como subcadena
    -5 si longitud < 8
    Puntaje mínimo 0.

    Ejemplos
    --------
    >>> password_score("abc123DEF")
    2 + 7 -3 -5 = 1
    >>> password_score("H0la#mundo")
    8 + 9 = 17
    """
    if not pwd:                       # cadena vacía → 0
        return 0

    score = 0

    # 1. Bonus por clases de caracteres
    classes = 0
    if any(c.isupper() for c in pwd):
        classes += 1
    if any(c.islower() for c in pwd):
        classes += 1
    if any(c.isdigit() for c in pwd):
        classes += 1
    if any(not c.isalnum() for c in pwd):
        classes += 1
    score += 2 * classes

    # 2. +1 por cada carácter único (que aparezca exactamente una vez)
    from collections import Counter
    freq = Counter(pwd)
    score += sum(1 for cnt in freq.values() if cnt == 1)

    # 3. Penalizaciones por subcadenas prohibidas (case-insensitive)
    lowered = pwd.lower()
    forbidden = ['123', 'abc', 'qwe']
    for seq in forbidden:
        if seq in lowered:
            score -= 3
            break   # solo se resta una vez

    # 4. Penalización por longitud
    if len(pwd) < 8:
        score -= 5

    # 5. Puntaje mínimo 0
    return max(score, 0)



print(password_score("abc123DEF")) 
print(password_score("H0la#mundo")) 
print(password_score("1234567"))     
print(password_score("Qwe"))         
print(password_score("aA1#"))        
```

```python
def fibonacci_inverso(n: int) -> int:
    """
    Devuelva el **índice** del primer término de la sucesión
    de Fibonacci (1,1,2,3,5,…) que **contenga** al número `n`
    como **subcadena** en su representación decimal.
    Si `n` es negativo, utilice el valor absoluto.

    Ejemplos
    --------
    >>> fibonacci_inverso(2)
    3   # F3 = 2
    >>> fibonacci_inverso(13)
    7   # F7 = 13
    >>> fibonacci_inverso(21)
    8   # F8 = 21
    >>> fibonacci_inverso(123)
    12  # F12 = 144 → no; F15 = 610 → no; F20 = 6765 → ¡sí! (contiene "123")
    """
    
    target = str(abs(n))
    a, b = 1, 1          # F1, F2
    idx = 1

    while True:
        if target in str(a):
            return idx
        a, b = b, a + b
        idx += 1

print(fibonacci_inverso(2))    # 3
print(fibonacci_inverso(13))   # 7
print(fibonacci_inverso(21))   # 8
print(fibonacci_inverso(123))  # 20
```
